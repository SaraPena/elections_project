```python
import pandas as pd
```

```python
with open('2018_senate_election_results.txt', 'r') as data:
        lines = data.readlines()
```

```python
type(lines)
```

```python
precinct=[]
rep=[]
dem=[]
lib=[]
over_votes=[]
under_votes=[]
```

```python
for line in lines:
    precinct.append(line[0:4])
    rep.append(line[37:41])
    dem.append(line[44:47])
    lib.append(line[51:53])
    over_votes.append(line[58:59])
    under_votes.append(line[63:65])
    
```

```python
senate_df=pd.DataFrame({'precinct':precinct,'rep':rep,'dem':dem,'lib':lib,'over_votes':over_votes,'under_votes':under_votes})
```

```python
senate_df.head()
```

```python
senate_df.apply(lambda x: x.str.strip())
```

```python
senate_df['rep']=pd.to_numeric(senate_df['rep'])
senate_df['dem']=pd.to_numeric(senate_df['dem'])
senate_df['lib']=pd.to_numeric(senate_df['lib'])
senate_df['over_votes']=pd.to_numeric(senate_df['over_votes'])
senate_df['under_votes']=pd.to_numeric(senate_df['under_votes'])
```

```python
senate_df.info()
```

```python
senate_df['total'] = senate_df['rep']+senate_df['dem']+senate_df['lib']
```
